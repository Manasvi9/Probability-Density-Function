```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/data.csv', encoding='latin-1')
df.head()
df
```

```python
x = df["no2"].dropna().values
#Step 1
r = 102317206
a_r = 0.05 * (r % 7)
b_r = 0.3 * (r % 5 + 1)
#Step 2
z = x + a_r * np.sin(b_r * x)
#Step 3
mu = np.mean(z)
sigma = np.mean((z - mu) ** 2)
lambdaa = 1 / (2 * sigma)
c = np.sqrt(lambdaa / np.pi)
print("Parameters:")
print("mu =", mu)
print("lambda =", lambdaa)
print("c =", c)
```

```python
pz = c * np.exp(-lambdaa * (z - mu) ** 2)
print("First 5 predicted probabilities:")
print(pz[:5])
```
